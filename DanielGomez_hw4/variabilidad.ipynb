```python
%pylab inline
from scipy.fftpack import fft,fftfreq
```

```python
datos=genfromtxt('RR_Lyrae_template.dat')

def nFrecuencias(datos,n):
    Fx=fft(datos)
    Fx[n:-n]=0
    return real(ifft(Fx))

def plotN(N=10):
    figure(figsize=(20,3*N))
    for i in arange(N)+1:
        subplot(N,1,i)
        plot(datos[:,0],nFrecuencias(datos[:,1],i))
        title('N='+str(i))
        
def plotChi2(datos,N):
    X=zeros(shape=(N,2))
    for i in range(N):
        X[i,0]=i+1
        X[i,1]=sum((nFrecuencias(datos,i+1)-datos)**2)
    plot(X[:,0],X[:,1])
    xlabel('N')
    ylabel('chi2')
```

```python
plotN()
```
