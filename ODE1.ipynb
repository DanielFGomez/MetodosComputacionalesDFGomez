    Try solving the same differential equation using all the methods, with a smaller time-step (say h=0.1). Which method is the worst?

    Solve the differential equation yâ€²=1+y2 in the interval 0<x<1 and the boundary condition y(0)=0. Use the four methods explained above. Compare the numerical solution against the analytical solution.



```python
h=0.01
min_x = 0.0
max_x = 1.0
n_points = int((max_x - min_x)/h)
x = zeros(n_points)
y = zeros(n_points)
def reset():
    global x
    global y
    x = zeros(n_points)
    y = zeros(n_points)
    x[0]=min_x
    y[0]=0
    
def fPrima(y):
    return 1.0+y**2

def mostrar():
    global x,y
    subplot(1,2,1)
    plot(x,y)
    subplot(1,2,2)
    plot(x,abs(y-tan(x)))
```

```python
for i in range(n_points):
    x[i]=x[i-1]+h
    y[i] = y[i-1] + h * fPrima(y[i-1])
mostrar()
```

```python
reset()

x[1] = min_x + h
y[1] = y[0] + h*fPrima(y[0])

for i in range(n_points):
    
    x[i] = x[i-1] + h
    y[i] = y[i-2] + 2 * h * fPrima(y[i-1])
    
mostrar()
```

```python
reset()
for i in range(n_points):
    xMedia=x[i-1]+0.5*h
    yMedia=y[i-1]+0.5*h*fPrima(y[i])
    
    x[i]=x[i-1]+h
    y[i]=y[i-1]+h*fPrima(yMedia)
    
mostrar()
```

```python
reset()
for i in range(n_points):
    xMedia=x[i-1]+0.5*h
    yMedia=y[i-1]+0.5*h*fPrima(y[i])
    
    x[i]=x[i-1]+h
    y[i]=y[i-1]+h*fPrima(yMedia)
    
mostrar()
```

```python
reset()
for i in range(n_points):
    k1=h*fPrima(y[i-1])
    k2=h*fPrima(y[i-1]+0.5*k1)
    k3=h*fPrima(y[i-1]+0.5*k2)
    k4=h*fPrima(y[i-1]+k3)
    x[i]=x[i-1]
    y[i]=y[i-1]+(k1+2*k2+2*k3+k4)
mostrar()
```
