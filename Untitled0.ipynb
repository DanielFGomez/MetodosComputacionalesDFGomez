```python
n=1000
def animate(i):
    r=(random.random(n))**(0.1*i)
    theta=2*pi*random.random(n)
    x=r*cos(theta)
    y=r*sin(theta)
    scatter(x,y,alpha=0.4)
    axis('equal')
    
animation.FuncAnimation(animate)
```

```python
from mpl_toolkits.mplot3d import Axes3D
n=1000
r=1
theta=(2*pi*random.random(n))
phi=arcsin(pi*random.random(n))
x=r*cos(theta)*sin(phi)
y=r*sin(theta)*sin(phi)
z=r*cos(phi)
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

o=fig.gca(projection='3d')
o.scatter(x,y,z)


o.set_xlabel('x')
o.set_ylabel('y')
o.set_zlabel('z')
```
