```python
datos=genfromtxt("munich.txt")
datos=delete(datos,where(abs(datos[:,1])>50),axis=0)
```

```python
x0=ones(len(datos))
x1=sin(datos[:,0]*2*pi)
x2=cos(datos[:,0]*2*pi)
mat=matrix(c_[x0,x1,x2])
y=matrix(datos[:,1]).T

par=linalg.inv(mat.T*mat)*mat.T*y
a=sqrt(par[1,0]**2+par[2,0]**2)
b=arctan2(-par[1,0],par[2,0])
c=par[0,0]

x=array(datos[:,0])
figure(figsize=(20,10))
yy=a*cos(2*pi*x+b)+c
yyy=array(mat*par)
scatter(datos[:,0],datos[:,1])
plot(x,yy,color="r",linewidth=5)


e=y-matrix(yy).T
error=(e.T*e/(len(e)-1.0))[0,0]
print "Error= " +str(error)

```

```python
x=array(datos[:,0])
x0=ones(len(datos))
x1=sin(x*2*pi)
x2=cos(x*2*pi)
x3=x
mat=matrix(c_[x0,x1,x2,x3])
y=matrix(datos[:,1]).T

par=linalg.inv(mat.T*mat)*mat.T*y
a=sqrt(par[1,0]**2+par[2,0]**2)
b=arctan2(-par[1,0],par[2,0])
c=par[0,0]
m=par[3,0]

figure(figsize=(20,10))
yy=a*cos(2*pi*x+b)+c+m*x
yyy=array(mat*par)
scatter(datos[:,0],datos[:,1])
plot(x,yy,color="r",linewidth=5)


e=y-matrix(yy).T
error=(e.T*e/(len(e)-1.0))[0,0]
print "Error= " +str(error)
print "m= " + str(m)
```

###### Se puede ver que incluyendo el termino del calentamiento global los datos se ajustan ligeramente mejor, esto es de esperarse ya que al aumentar la cantidad de parametros siempre mejora el ajuste pero dado que tanto m como la diferencia de error es muy pequeña se puede decir que la evidencia de calentamiento no es muy fuerte.
