```python
n=128
f=200.0
dt=1/(f*32)
t=linspace(0,(n-1)*dt,n)
y=cos(2*pi*f*t)
```

```python
plot(t,y)
plot(t,y,"ko")
xlabel("time(s)")
ylabel("y(t)")
```

```python
y=cos(2*pi*f*t)-0.4*sin(2*pi*(2*f)*t)
plot(t,y)
plot(t,y,"ko")
xlabel("time(s)")
ylabel("y(t)")
```

```python
from scipy.fftpack import fft,fftfreq
fft_x=fft(y)/n #Esto es para la normalizacion, es importante saber como normaliza la funcion importada ya que son diferentes
freq=fftfreq(n,dt)
bar(freq,abs(fft_x),width=57
```
