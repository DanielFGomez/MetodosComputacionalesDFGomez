```python
%pylab inline
from scipy.fftpack import fft,fftfreq
```

```python
colores=('Reds','Greens','Blues')

def pasaBajas(imagen,cutoff=12, w=5):
    
    x=imread(imagen)
    color=False
    if(len(shape(x))==3):
        Sx,Sy=shape(x[:,:,0])
        color=True
    else:
        Sx,Sy=shape(x)
    
    #Define la mascara que implementa un filtro suave pasa bajas
    mask=zeros(shape=[Sx,Sy])
    for i in arange(Sx)-Sx/2:
        for j in arange(Sy)-Sy/2:
            if(sqrt(i**2+j**2)<cutoff-w):
                #1 para frecuencias bajas
                mask[i,j]=1
            elif(sqrt(i**2+j**2)<cutoff+w):
                #Transicion suave
                mask[i,j]=0.5*(1-sin(pi*(sqrt(i**2+j**2)-cutoff)/(2.0*w)))
    #Para imagenes a color se devuelve por separado el filtro para cada color
    if(color):
        NewImg=zeros(shape=shape(x))
        for i in range(3):
            NewImg=ifft2(fft2(x[:,:,i])*mask)
            subplot(2,2,i)
            imshow(abs(NewImg), cmap=colores[i])
    else:
        NewImg=ifft2(fft2(x)*mask)
        imshow(abs(NewImg), cmap='gray')
 
def pasaAltas(imagen,cutoff=12, w=5):
    x=imread(imagen)
    color=False
    if(len(shape(x))==3):
        Sx,Sy=shape(x[:,:,0])
        color=True
    else:
        Sx,Sy=shape(x)
    
    #Define la mascara que implementa un filtro suave pasa altas
    mask=ones(shape=[Sx,Sy])
    for i in arange(Sx)-Sx/2:
        for j in arange(Sy)-Sy/2:
            if(sqrt(i**2+j**2)<cutoff-w):
                #0 para frecuencias bajas
                mask[i,j]=0
            elif(sqrt(i**2+j**2)<cutoff+w):
                #transicion suave
                mask[i,j]=0.5*(sin(pi*(sqrt(i**2+j**2)-cutoff)/(2.0*w)))
    #Para imagenes a color se devuelve por separado el filtro para cada color
    if(color):
        NewImg=zeros(shape=shape(x))
        for i in range(3):
            NewImg=ifft2(fft2(x[:,:,i])*mask)
            subplot(2,2,i)
            imshow(abs(NewImg), cmap=colores[i])
    else:
        NewImg=ifft2(fft2(x)*mask)
        imshow(abs(NewImg), cmap='gray')
    
    
```

```python
pasaBajas('full_moon.jpg')
```

```python
pasaAltas('full_moon.jpg')
```

```python
imagenes=('20_popc_cho009-1.tif','colesterol-1.tif','ves_full_150_002-1.tif')
j=1
figure(figsize=(20,20))
for i in imagenes:
    subplot(3,3,j)
    imshow(imread(i),cmap='gray')
    j+=1
    subplot(3,3,j)
    pasaBajas(i)
    j+=1
    subplot(3,3,j)
    pasaAltas(i,w=9)
    j+=1
```

Cualitativamente hablando el filtra pasa bajas hace mas borrosa la imagen, el pasa altas resalta los contrastes en la imagen.
